## Linear Regression Model

```python
import pandas as pd 
import numpy as np
from sklearn.linear_model import  LinearRegression
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\AVANISH SINGH\OneDrive\Desktop\Advertising.csv")
```

```python
x = df[['TV', 'Radio', 'Newspaper']]   # Features are independent variables
# in x we are using TV, Radio, and Newspaper as features.
print(x)
```

```python
y = df['Sales']
# in y we are using Sales as target variable.
print(y)
```

```python
from sklearn.linear_model import LinearRegression

X = df[['TV', 'Radio', 'Newspaper']]
y = df['Sales']

model = LinearRegression()
model.fit(X, y)

print("Coefficients:", model.coef_)
print("Intercept:", model.intercept_)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plot TV vs Sales with regression line
sns.regplot(x='TV', y='Sales', data=df)
plt.title("TV vs Sales with Regression Line")
plt.show()

# Plot Radio vs Sales with regression line
sns.regplot(x='Radio', y='Sales', data=df)
plt.title("Radio vs Sales with Regression Line")
plt.show()

# Plot Newspaper vs Sales with regression line
sns.regplot(x='Newspaper', y='Sales', data=df)
plt.title("Newspaper vs Sales with Regression Line")
plt.show()
```

```python
model = LinearRegression()
# importing the necessary libraries for linear regression.
```

```python
model.fit(x, y)
# model.fit is use to train the model. and it fits the model.
```

```python
from sklearn.model_selection import train_test_split # for splitting the data into training and testing.
```

```python
from sklearn.linear_model import LinearRegression # for linear regression.
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y,test_size=0.2, random_state=42) 
# splitting the data into training and testing set. 
```

```python
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error
MAE = mean_absolute_error(y_test, y_pred)
MAE
```
