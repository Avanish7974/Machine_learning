```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\AVANISH SINGH\OneDrive\Desktop\Excel_machine\spam.csv",encoding='latin-1')
```

```python
x=df['v2']
y=df['v1']
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
bow_clf=TfidfVectorizer(max_features=20,ngram_range=(2,4))
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train1=bow_clf.fit_transform(x_train)
```

```python
x_test1=bow_clf.transform(x_test)
from sklearn.svm import SVC
model = SVC(kernel='linear')
model.fit(x_train1,y_train)
```

```python
y_pred=model.predict(x_test1)
```

```python
from sklearn.metrics import accuracy_score
# print(confusion_matrix(y_test, y_pred))
# print(classification_report(y_test, y_pred))
print(accuracy_score(y_test, y_pred))
```
