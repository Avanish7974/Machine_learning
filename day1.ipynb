```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
data = {'country' : ['Hunngary','korea','France','Australia','US'],
        'GDP' : [12240,27195,37675,50962,55805],
        'Life':[4.9,5.8,6.5,7.3,7.2]
        }
df = pd.DataFrame(data)
```

```python
import seaborn as sns
sns.scatterplot(x='GDP',y='Life',data = df)
```

```python
model = LinearRegression()
```

```python
model.fit(df[['GDP']],df['Life'])
# this shows the parameters of the linear regression line. 
```

```python
df
```

```python
model.predict([[1000]])
```

```python
model.predict([[52000]])
```

```python
# in the equation of line 
# y = mx + c
# y = is the independent variable,
# m is the coefficient of x, and m is the slope of the line. and tengent  of the linear regression line.
# c is the intercept. In this case, 
# y is Life, m is 0.0191, and 
# c is 4.
```

```python
model.coef_  # m
# it shows the coefficient of GDP
```

```python
model.intercept_  # c
# it shows the intercept of the line
```

```python
# by the above coef and intercept we can draw the line for the given data. 
sns.regplot(x='GDP',y='Life',data = df)
```
