## StandardScaler

```python
from sklearn.preprocessing import StandardScaler
import numpy as np

X_train = np.array([[1], [2], [3], [4], [5]])   # training
X_test  = np.array([[6], [7]])                  # test

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)   #  training
X_test_scaled  = scaler.transform(X_test)        #  test

print(X_train_scaled)
print(X_test_scaled)
```

## CountVectorizer



```python
from sklearn.feature_extraction.text import CountVectorizer

train_texts = ["I love machine learning", "Machine learning is fun"]
test_texts  = ["I love deep learning"]

vectorizer = CountVectorizer()

X_train = vectorizer.fit_transform(train_texts)   #  training
X_test  = vectorizer.transform(test_texts)        #  test

print("Vocab:", vectorizer.vocabulary_)
print("Train:", X_train.toarray())
print("Test:", X_test.toarray())
```
