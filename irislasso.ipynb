```python
##25/07/2025
##standard scaler is used to enhance the accuracy of the model
##in standard scalar we normalise the data by two methods 1. z- score(-1 to 1) with outliers 2> min-max(0-1) normalisation through min-max scalar with no or few outliers
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge,LinearRegression,Lasso
from sklearn.preprocessing import StandardScaler


```

```python
df=pd.read_csv('/content/iris.csv')
df
```

```python
from sklearn import preprocessing

label_encoder = preprocessing.LabelEncoder()
df['species']= label_encoder.fit_transform(df['species'])

```

```python

df
```

```python
x= df[['sepal_length'	,'sepal_width'	,'petal_length','petal_width']]
y = df['species']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
x_train
```

```python
model = Lasso(alpha=0.1)
model.fit(x_train, y_train)


```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score,confusion_matrix,classification_report
```

```python
MSE=mean_squared_error(y_test,y_pred)
MSE
```

```python
MAE=mean_absolute_error(y_test,y_pred)
MAE
```

```python
r2_score=r2_score(y_test,y_pred)
r2_score
```
