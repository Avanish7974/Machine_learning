```python
x =df['v2']
y =df['v1']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
bow_clf = CountVectorizer(max_features=20,stop_words='english')
# stop_words = english means the common words in english language like is,am,are etc
```

```python
bow_clf
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train1=bow_clf.fit_transform(x_train)
```

```python
x_train1
```

```python
x_train1.toarray()
```

```python
x_test1=bow_clf.transform(x_test)
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
```

```python
model.fit(x_train1,y_train)
```

```python
y_pred = model.predict(x_test1)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
print(accuracy_score(y_test,y_pred))
```

```python
print(confusion_matrix(y_test,y_pred))
```
